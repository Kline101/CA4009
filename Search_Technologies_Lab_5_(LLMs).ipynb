#Large-Language Model (LLMs)
Next we load a **Flan-T5** Language model. This model has been hosted in HuggingFace. Note that we are using a *base* variant of the T5 model for this lab.

#Method
We will be using the *pipeline()* method. The following block uses the *pipeline* method to specify the task and the model we will use for the excersie.

Further details on the *pipeline()* method can be found [here](https://huggingface.co/docs/transformers/v4.35.2/en/main_classes/pipelines#transformers.pipeline).

```python
pipe = pipeline(
    "text2text-generation",
    model=model,
)
```

Recently, large language models (LLMs) have shown amazing capabilities for text generation and conversation understanding. In the field of information retrieval (IR), LLMs have also been successfully utilized to enhance relevance modeling via various techniques such as query generation, query expansion, document prediction, etc.

Specifically, we first prompt
LLM to generate query rewrites.

 We design and explore three prompting methods. The methods will use the original query and Relevant document(s) from the IRModelGenerator used in the previous labs.

 We will prompt the large-language model for Query rewriting on 3 different scenarios:

 1. Zero-shot
 2. One-shot
 3. Few-shot


#Zero-shot

In the absence of any specific document or context, the zero-shot scenario entails generating a search query solely based on the given query. Given original query *'International Organized Crime.'* A potential rewritten query could be:

"Recent global efforts to combat international organized crime."

```python
prompt = """Rewrite a better search query for a web search engine given Query.
Query: "International Organized Crime"
Rewrite:"""


sequences = pipe(
    prompt,
)


for seq in sequences:
    print(f"Rewrite: {seq['generated_text']}")
```
